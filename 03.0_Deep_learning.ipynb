To visualize our model's predictions we're going to create a function plot_decision_boundary() which:
1->Takes in a trained model, features (X) and labels (y).
2->Creates a meshgrid of the different X values.
3->Makes predictions across the meshgrid.
4->Plots the predictions as well as a line between the different zones (where each unique class falls)

```python


def plot_decison_boundary(model,X,y):
    x_min,x_max = X[:,0].min()-0.1 , X[:,0].max()+0.1
    y_min,y_max = X[:,1].min()-0.1 , X[:,1].max()+0.1
    xx , yy = np.meshgrid(np.linspace(x_min,x_max,100),
                         np.linspace(y_min,y_max,100))
    
    x_in = np.c_[xx.ravel(), yy.ravel()] # stack 2D array together
    
    y_pred = model.predict(x_in)
    
    # check for multiclass
    
    if len(y_pred[0]) > 1:
        print("doing a multiclass classification")
        # we have to reshape our prediction to get them ready for plotting
        y_pred = np.argmax(y_pred , axis=1).reshape(xx.shape)
        
    else:
        print("Doing a binary classification")
        y_pred = np.round(y_pred).reshape(xx.shape)
        
    #plot the decision boundary
    plt.contourf(xx ,yy ,y_pred ,cmap=plt.cm.RdYlBu,alpha=0.7)
    plt.scatter(X[:,0],X[:,1], c=y,s=40 ,cmap=plt.cm.RdYlBu)
    plt.xlim(xx.min() ,xx.max())
    plt.ylim(yy.min() ,yy.max())
```

```python
plot_decison_boundary(model=model_2,
                     X=X,
                     y=y)
```

# Finding best learning rate

```python
model_6 = tf.keras.Sequential([
    tf.keras.layers.Dense(4,activation="relu"),
    tf.keras.layers.Dense(4,activation="relu"),
    tf.keras.layers.Dense(1,activation="sigmoid"),
])

model_6.compile(loss=tf.keras.losses.BinaryCrossentropy(),
                optimizer=tf.keras.optimizers.Adam(),
                metrics=["accuracy"])

lr_scheduler = tf.keras.callbacks.LearningRateScheduler(lambda epoch: 1e-4 * 10**(epoch/20))

history_1=model_6.fit(X_train,y_train,epochs=100,callbacks=[lr_scheduler])
```

```python
pd.DataFrame(history_1.history).plot()
```

```python
lrs = 1e-4 * (10 ** (tf.range(100) / 20))
plt.figure(figsize=(12,8))
plt.semilogx(lrs, history_1.history["loss"])
plt.xlabel("learning rate")
plt.ylabel("loss")
plt.title("learning rate vs loss")
```
